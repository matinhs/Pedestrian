```python
model1 = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/best1.pt')  
model2 = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/best2.pt')  
```

Test set with model 1

```python
from os.path import isfile, join
pathIn= '/content/Test/'
pathOut = 'Test_1.mp4'
fps = 10
frame_array = []
files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
#for sorting the file names properly
files.sort(key = lambda x: x[5:-4])
files.sort()
frame_array = []
files = [f for f in os.listdir(pathIn) if isfile(join(pathIn, f))]
#for sorting the file names properly
files.sort(key = lambda x: x[5:-4])
for i in range(len(files)):
    filename=pathIn + files[i]
    #reading each files
    img = cv2.imread(filename)
    results = model1(img[:,:,::-1])
    x = results.pandas().xyxy[0]
    if len(x) != 0:
      for j in range(len(x)):
        row = x.iloc[j]
        img = cv2.rectangle(img,(int(row[0]),int(row[1])),(int(row[2]),int(row[3])),(0,255,0),1)
        img = cv2.putText(img, str(row[6])+' '+str(row[4])[:3], (int(row[0]),int(row[1])-3), cv2.FONT_HERSHEY_PLAIN, 0.6, (0,0,0))

    height, width, layers = img.shape
    size = (width,height)
    
    #inserting the frames into an image array
    frame_array.append(img)
out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
for i in range(len(frame_array)):
    # writing to a image array
    out.write(frame_array[i])
out.release()
```
